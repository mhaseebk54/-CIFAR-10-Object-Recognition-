```python
from tensorflow.keras import Sequential,layers,models
from tensorflow.keras.layers import Flatten,Dense,Dropout
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras import optimizers
```

```python
convontial_base = ResNet50(weights='imagenet',include_top=False,input_shape=(256,256,3))
convontial_base.summary()
```

```python
numofclass = 10

model = models.Sequential()
model.add(layers.UpSampling2D((2,2)))
model.add(layers.UpSampling2D((2,2)))
model.add(layers.UpSampling2D((2,2)))
model.add(convontial_base)
model.add(layers.Flatten())
model.add(layers.BatchNormalization())
model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.BatchNormalization())
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(numofclass,activation='softmax'))
```

```python
model.compile(optimizer=optimizers.RMSprop(learning_rate=2e-5),loss='sparse_categorical_crossentropy',metrics=['acc'])
```

```python
history = model.fit(X_train_scale,Y_train,validation_split=0.1,epochs=10)
```

```python
loss,accuracy = model.evaluate(X_test_scale,Y_test)
print(accuracy)
```
